```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
MyGraph = nx.Graph()

MyGraph.add_node('A', npos=(0, 0), ccn='blue')
MyGraph.add_node('B', npos=(4, 4), ccn='blue')
MyGraph.add_node('C', npos=(11, 4), ccn='blue')
MyGraph.add_node('D', npos=(15, 4), ccn='blue')
MyGraph.add_node('E', npos=(17, 6), ccn='blue')
MyGraph.add_node('F', npos=(21, 10), ccn='blue')

MyGraph.add_edge('A', 'B', cce='blue', label='0.5 km')
MyGraph.add_edge('B', 'C', cce='blue', label='0.7 km')
MyGraph.add_edge('C', 'D', cce='blue', label='0.4 km')
MyGraph.add_edge('D', 'E', cce='blue', label='0.3 km')
MyGraph.add_edge('E', 'F', cce='blue', label='0.5 km')

pos = nx.get_node_attributes(MyGraph, 'npos')
nodecolour = nx.get_node_attributes(MyGraph, 'ccn')
edgecolour = nx.get_edge_attributes(MyGraph, 'cce')

nodearray = nodecolour.values()
edgearray = edgecolour.values()

plt.figure(figsize=(21, 10))

plt.text(0.27, -0.07, s='Hyde Park Corner', rotation=0)
plt.text(4, 4.27, s='Green Park', rotation=15)
plt.text(11, 4.27, s='Piccadilly Circus', rotation=15)
plt.text(14.4, 3.3, s='Leicester Square', rotation=15)
plt.text(17.27, 5.93, s='Covent Garden', rotation=0)
plt.text(19.9, 9.93, s='Holborn', rotation=0)

nx.draw_networkx(MyGraph, pos, node_color=nodearray)
nx.draw_networkx_edges(MyGraph, pos, edge_color=edgearray, width=5)
edge_labels = nx.get_edge_attributes(MyGraph, 'label')
nx.draw_networkx_edge_labels(MyGraph, pos, edge_labels=edge_labels)

plt.show()
```
